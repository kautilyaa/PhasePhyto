# PhasePhyto Run Results Inspector -- Split Workflow

This former all-in-one inspector has been split into focused notebooks:

- `PhasePhyto_Inspect_00_Index.ipynb` -- list runs and choose an inspection path.
- `PhasePhyto_Inspect_01_Run_Overview.ipynb` -- manifest, artifacts, available runs.
- `PhasePhyto_Inspect_02_Metrics.ipynb` -- metrics and PhasePhyto-vs-baseline comparison.
- `PhasePhyto_Inspect_03_Plots.ipynb` -- plots and analysis sample images.
- `PhasePhyto_Inspect_04_Reports.ipynb` -- target classification report and compact summary.

Open `PhasePhyto_Inspect_00_Index.ipynb` first.
