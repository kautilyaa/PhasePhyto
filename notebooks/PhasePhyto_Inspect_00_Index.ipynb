```python
# List available runs so you can choose CONFIG["run_name"] in the focused notebooks.
from pathlib import Path
from google.colab import drive

drive.mount("/content/drive", force_remount=False)
PROJECT_DIR = Path(CONFIG["drive_project_dir"])
RUNS_DIR = PROJECT_DIR / "runs"

if not RUNS_DIR.exists():
    raise RuntimeError(f"No runs directory found: {RUNS_DIR}")

runs = sorted([p for p in RUNS_DIR.iterdir() if p.is_dir()])
print(f"Found {len(runs)} run(s) under {RUNS_DIR}")
for run in runs:
    print(f"  {run.name}")

if runs:
    print(f"\nLatest run: {runs[-1].name}")
```
